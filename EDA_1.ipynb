```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(3)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
b = final_df.Country.value_counts()
print(b)
print(b.values)
c = b.values
print("sum =" ,b.sum())
```

```python
#Pie-Chart - top 3 countries that uses zomato
plt.pie(c[:3],autopct="%1.2f" )
```

```python
final_df.columns
```

```python
c = final_df.groupby(['Aggregate rating','Rating color','Rating text' ]).size()
print(c)
c = c.reset_index().rename(columns={0:'Rating Count'})
c
```

```python
# Observations ---> Rating bwt 4.5 to 4.9 = Excellent
#              ---> Rating bwt 4.0 to 4.4 = Very Good
#              ---> Rating bwt 3.5 to 3.9 = Good
#              ---> Rating bwt 3.0 to 3.4 = Very Good
#              ---> Rating bwt 2.5 to 2.9 = average
#              ---> Rating bwt 4.0 to 2.4 = poor
```

```python
c.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=c,palette=['black','red','yellow','green','green'])
```

```python
#observation ----> not rated is very high
#            ----> max ratings are from 2.5 to 4.4
```

```python
sns.countplot(x='Rating color', data=c,palette=['black','red','yellow','green','green'])
```
